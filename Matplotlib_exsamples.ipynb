We need to lsts[6], and lists[10]

```python
df = lists[6]
df.info()
```

```python
df.head()
```

```python
df.set_index('Country', inplace=True)
```

```python
df.sort_values('Athletes', inplace=True)
```

```python
plt.figure(figsize=(64,30))
plt.title("2016 year olimpics Athletes")
plt.xticks(rotation=90)
sns.barplot(x=df.IOC, y=df.Athletes)
plt.show()
```

```python
df[df.IOC == 'UZB']
```

```python
df1 = df[df.Athletes > 200]
```

```python
plt.figure(figsize=(18,10))
plt.title('Athletes namber')
plt.xlabel("Country cod")
plt.ylabel('Number')
plt.xticks(rotation=90)
sns.barplot(x=df1.IOC, y=df1.Athletes)
plt.show()
```
