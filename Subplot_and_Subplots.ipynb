# Subplot_exsamples

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
a = -5
b = 0
c = 2 
x = np.arange(-4,6,0.4)

kvadrat_tengsizlik = a * x * x + b * x + c

plt.figure(figsize=(12,6))
plt.plot(kvadrat_tengsizlik,
         linewidth=1.5,
         linestyle='-',
         color='b',
         alpha=0.9
         )
plt.title("Kvadrat tenglama grafigi")
plt.xlabel("x o'qi")
plt.ylabel("y o'qi")
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
x = np.arange(-4,6,0.4)
kubik_tengsizlik = x * x * x
plt.figure(figsize=(12,6))
plt.plot(kubik_tengsizlik,
         linewidth=1.5,
         linestyle='-.',
         color='r',
         alpha=0.9)
plt.title("Kubik_tenglama grafigi")
plt.xlabel("x o'qi")
plt.ylabel("y o'qi")
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

subplot

```python
a = -5
b = 0
c = 2 
x = np.arange(-4,6,0.4)

kvadrat_tengsizlik = a * x * x + b * x + c

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(kvadrat_tengsizlik,
         linewidth=1.5,
         linestyle='-',
         color='b',
         alpha=0.9
         )
plt.title("Kvadrat tenglama grafigi")
plt.xlabel("x o'qi")
plt.ylabel("y o'qi")
plt.grid(which='both')
plt.minorticks_on()

kubik_tengsizlik = x * x * x

plt.subplot(1,2,2)
plt.plot(kubik_tengsizlik,
         linewidth=1.5,
         linestyle='-.',
         color='r',
         alpha=0.9)
plt.title("Kubik_tenglama grafigi")
plt.xlabel("x o'qi")
plt.ylabel("y o'qi")
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

# Seaborn and subplots

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

`subplots` funksiyasi 2 ta obyekt qaytaradi:
1. Umumiy chizma obyekti
2. Chizma tarkibidagi grafiklar ro'yxati

Biz quyida, qulaylik uchun, bu ikkalasini `fig` (figure) va `ax` (axes) o'zgaruvchilariga saqlab olamiz:

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

fig.suptitle('grafig 1')
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x=df['year'], y=df['price'], hue='fuelType', s=90)
plt.show()
```

2 o'lchamli grafik

```python
fig, ax = plt.subplots(2,2, figsize=(18,10))
fig.suptitle("grafik 2")
sns.barplot(ax=ax[0,0], data=df, x=df['year'], y=df['price'])
sns.scatterplot(ax=ax[0,1], data=df, x=df['year'], y=df['mileage'], hue=df['fuelType'])
sns.scatterplot(ax=ax[1,0], data=df, x=df['year'], y=df['engineSize'], hue=df['transmission'])
sns.histplot(ax=ax[1,1], data=df, x=df['year'], y=df['price'])

ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("EngineSize vs Year correlation")
ax[1,1].set_title("Price vs Year correlation")
plt.show()
```
