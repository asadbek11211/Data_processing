# About seaborn library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
sns.lineplot(data=df[['Firehouse Museum','Avila Adobe']])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Muzeyga tashriflar soni")
sns.lineplot(data=df)
```

```python
plt.figure(figsize=(17,8))
plt.title("Muzeyga tashrf buyuruvchilar soni")
sns.lineplot(data=df)
plt.xlabel('Sana')
plt.xticks(rotation=90)
plt.ylabel('Odamlar soni')
plt.show()
```

**The main task of line graphs is to monitor the change (trend) of a process/event over time**

From the above graph, we can see the number of visitors to different museums in a beautiful and, most importantly, understandable way and draw different conclusions.

For example, Avila Adobe Museum is very famous among museums. And the Firehouse Museum was overcrowded on January 9, 2014 (this may also indicate an error in the schedule)
