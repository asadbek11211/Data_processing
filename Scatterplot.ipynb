# Scatter plot-tarqoqlik grafigi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='year', y='price')
plt.ylabel("Aholi soni")
plt.grid()
plt.show()
```

```python
df = df[df.year > 2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.ylabel("Aholi soni")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission')
plt.ylabel("Aholi soni")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=120)
plt.ylabel("Aholi soni")
plt.grid()
plt.show()
```
