# *Heatmap*

The table below shows the number of visitors to some museums in the United States on different dates. Do we only read the values for the first museum ( usecols=[0,1] )

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

We divide the data by year and month.

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df.groupby(['month','year']).max().head(15)
```

```python
df1 = df.groupby(['month','year']).max().unstack(level=0)
```

```python
df1
```

```python
plt.figure(figsize=(12,6))
plt.title('Avila adobe muzeyi')
plt.xlabel('month')
plt.ylabel('yil')
plt.grid()
sns.heatmap(data=df1)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(12,6))
plt.title('Avila adobe muzeyi')
plt.xlabel('month')
plt.ylabel('yil')
plt.grid()
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
sns.heatmap(data=df1, cmap='YlGnBu', linewidth=1.1, square=True)
plt.show()
```
