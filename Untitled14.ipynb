```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df)
```

```python
sns.histplot(data=df, x=df.level, y=df.price)
plt.show()
```

```python
sns.barplot(data=df, x=df.level, y=df.price)
plt.show()
```

```python
sns.scatterplot(data=df, x=df.level, y=df.price)
plt.show()
```

```python
sns.relplot(data=df, x=df.level, y=df.price)
plt.show()
```

```python
sns.regplot(data=df, x=df.level, y=df.price)
plt.show()
```
