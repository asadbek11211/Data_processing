```python
import numpy as np
import re
from scipy.ndimage import zoom
import matplotlib.pyplot as plt


def ks(L, name):
    def forward(z, R2, R1, z0):
        # print(type(z))
        # print(z)
        # if type(z) == type(np.arange(1, 10, 1)):
        #     z = np.linspace(z[-1] / 2, -(z[-1] / 2), 200)
        #     print(z)
        H = (z - z0) / L
        K12 = (R2 - R1) / (R1 + R2)
        K21 = (2 * R1) / (R1 + R2)

        if z < -L:
            if R2 > R1:
                P = (1 + (K21 / (2 * H + 1) ** 2)) * R2
                return P
            elif R1 > R2:
                P = (1 + (K21 / (2 * H - 1) ** 2)) * R2
                return P
        elif z > L:
            if R2 > R1:
                P = R1 * (1 - (K12 / ((2 * H) - 1) ** 2))
                return P
            elif R1 > R2:
                P = R1 * (1 + (K12 / ((2 * H) + 1) ** 2))
                return P
        elif z < L:
            P = 2 * (R1 * R2) / (R1 + R2)
            return P
        elif z > -L:
            P = 2 * (R1 * R2) / (R1 + R2)
            return P

    vforward = np.vectorize(forward)

    vforward.__doc__ = name
    return vforward

L = [1, 0.3] #длина зонда
lfd = ks(L=2, name="Long zond")
lfs = ks(L=1, name="Short zond")

ftool = [lfd, lfs]

R2 = 100 #сопротивление верхнего пласта
R1 = 200 #сопротивление нижнего пласта
L = 5 #длина зонда
H = 10 #глубина границы двух слоев
#
#
z0 = 0
#
z = np.linspace(H, -H, 200)
print(z)
zreal = np.arange(H, -H, -1)
plt.figure(figsize=(15,10))
for zond in ftool:
     plt.plot(zond(z, R1, R2, z0), z, label=zond.__doc__)
#
#
ax = plt.gca()
ax.invert_yaxis()
plt.legend()
plt.show()
```
