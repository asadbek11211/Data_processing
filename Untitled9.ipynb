```python
import numpy as np
from scipy.special import erf


def generate_fake_zond(length, name):
    def forward(z, rho1, rho2, z0):
        h = z - z0
        d = (1 + erf(h / length / 2)) / 2
        rok = rho2 * d + rho1 * (1 - d)

        return rok

    vforward = np.vectorize(forward)

    vforward.__doc__ = name
    return vforward


lfd = generate_fake_zond(length=1, name="Long fake zond")
lfs = generate_fake_zond(length=0.3, name="Short fake zond")

ftool = [lfd, lfs]

rho1 = 10.0
rho2 = 200.0
z = np.arange(0, 101, 1)
z0 = 50

for zond in ftool:
     plt.plot(z, zond(z, rho1, rho2, z0), label=zond.__doc__)

plt.legend()
plt.show()









```
