# Chiziqli grafiklar. matplotlib paketi.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sonlar = np.arange(20)
sonlar
```

```python
plt.plot(sonlar)
plt.show()
```

Biroz murakkabroq misol. $sin(x)$ funksiyasining $0$ va $4\pi$ oralig'idagi grafigini chizamiz.

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.log(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, linewidth=3, linestyle='--')
plt.show()
```

Grafikga biroz o'zgartirish kiritamiz

```python
plt.plot(x,y,
         linewidth=2.8,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.9,
         marker='^'
         )
plt.show()
```

```python
plt.plot(x,y,
         linewidth=1.4,
         linestyle=':',
         color='b',
         alpha=0.9,
         marker='*'
         )
# grafik nomi:
plt.title('log(x) funksiya grafigi')
# x o'q nomi
plt.xlabel("x son o'qi")
# y o'qi nomi
plt.ylabel("y son o'qi")
# setka qo'shish
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y,
         linewidth=1.4,
         linestyle=':',
         color='b',
         alpha=0.9,
         marker='*'
         )
# grafik nomi:
plt.title('log(x) funksiya grafigi')
# x o'q nomi
plt.xlabel("x son o'qi")
# y o'qi nomi
plt.ylabel("y son o'qi")
# setka qo'shish
plt.grid()
plt.show()
```

Rasmni saqlash

```python
plt.figure(figsize=(10,4))
plt.plot(x,y,
         linewidth=1.4,
         linestyle=':',
         color='b',
         alpha=0.9,
         marker='*'
         )
# grafik nomi:
plt.title('log(x) funksiya grafigi')
# x o'q nomi
plt.xlabel("x son o'qi")
# y o'qi nomi
plt.ylabel("y son o'qi")
#har bir chiziq haqida ma'lumot
plt.legend(['log(x)'], loc=2)
# setka qo'shish
plt.grid()
plt.show()
#Rasmni sqlash
plt.savefig('log(x)_funksiyasi.png')
```
