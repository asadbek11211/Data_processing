```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/global_earthquake.csv?raw=true")
df.head()
```

Jadvalning `Date` ustuni zilzila sanasi,  `Time` ustuni esa zilzila vaqtini saqlaydi. Keling bu ustunlardagi ma'lumotlar turini ko'ramiz:

```python
df[['Date','Time']].info()
```

Date va Time obekt deb korsatdi, biz esa ularni datetime kutubxonasidan yotdalanib boshqa farmatga o'tqazamiz yani DateTime farmatiga

```python
df[["Date"]].head()
```

Bizda xato chiqardi sababi bunday farmat mavjud emas deyabdi hozir tekshirib ko'ramiz

```python
pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
df['Dates'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
df.head()
```

```python
df['month'] = df.Dates.dt.month
df['month'].head()
```

```python
df['Year'] = df.Dates.dt.year
df['Year']
```

```python
df['days'] = df.Dates.dt.day
df['days']
```
