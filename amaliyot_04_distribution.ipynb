## 4-AMALIYOT. Distribution


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

#### Yuqoridagi jadvalda Toshkent shahridagi sotildagian uylar haqida ma'lumotlar jamlangan.

### 1-VAZIFA. Uylarning narxi taqsimotini grafikda ko'rsating

```python
#JAVOBINGIZNI SHUYERGA YOZING
sns.displot(data=df, x='price', height=5, aspect=2)
plt.show()
```

### Jadvalda 7000 dan oshiq uy bor, aksar uylarning narxi 200000$ dan kam. Keling noodatiy katta qiymatlarni tashlab yuboramiz va yuqoridagi chizmani qaytadan chizamiz.

```python
#JAVOBINGIZNI SHUYERGA YOZING
df = df[df['price'] < 200000]
sns.displot(data=df, x='price', height=4, aspect=2)
plt.show()
```

### 2-VAZIFA. Uylarning maydoni taqsimoti grafigini chizing

```python
# JAVOB UCHUN JOY
sns.displot(data=df, x='size', height=4, aspect=2)
plt.show()
```

### 3-VAZIFA. Uylarning xonalar soni taqsimotini chiqaring.

```python
# JAVOB UCHUN JOY
sns.displot(data=df, x='rooms', height=4, aspect=2)
plt.show()
```

6 xonadan ortiq uylarni tashlab yuboramiz.

```python
df = df[df['rooms']<7]
```

### 4-VAZIFA. Uylarning narxi taqsimotini xonalar soniga bog'lab chiqaramiz.

```python
# JAVOBNI SHUYERGA YOZING
sns.displot(data=df, x='price', hue='rooms', height=4, aspect=2)
plt.show()
```

### 4.1 Grafik tushunarsiz chiqdi. Keling faqatgina 1,2,3 xonali uylarning narx bo'yicha taqsimotini chizamiz.

```python
# JAVOBNI SHUYERGA YOZING
df1 = df[df['rooms'] < 4]
sns.displot(data=df1, x='price', hue='rooms', height=4,aspect=2)
plt.show()
```

### 5-VAZIFA. 4,5,6 xonali uylar uchun narx taqsimotini alohida grafiklarda chizing

```python
# JAVOB UCHUN JOY


sns.displot(data=df[df['rooms'] > 3], x='price', col='rooms', height=4, aspect=2)
plt.show()
```

## 6-VAZIFA. Uylarning maydoni taqsimoti KDE grafigini chizing

```python
# JAVOB UCHUN JOY

sns.displot(data=df, kind='kde', x='size', height=4, aspect=2)
plt.show()
```

## 7-VAZIFA. Uylarning maydoni taqsimotini KDE va taqsimotini birlashtirib chizing.

```python
# JAVOB UCHUN JOY

sns.displot(data=df, kde=True, x='size', height=4, aspect=2)
plt.show()
```

## 8-VAZIFA. Uylarning narxi va xonalari soni uchun (2,3,5 xonali) KDE chizing.

```python
# JAVOB UCHUN JOY
df3=df[(df['rooms'] == 2) | (df['rooms'] == 3) | (df['rooms'] == 5)]
sns.displot(data=df3, kind='kde', x='price', hue='rooms', height=4, aspect=2)
plt.show()
```

## 9-VAZIFA. Uylarning maydoni uchun kumulyativ taqsimot grafigini chizing (CDF)

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(8,4))
sns.ecdfplot(data=df, x='size')
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

## 10-VAZIFA. Yuqoridagi grafikni uylarning narxi uchun chizing.

```python
# JAVOB UCHUN JOY

sns.displot(data=df, x='price', kind='ecdf', height=6, aspect=2)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
