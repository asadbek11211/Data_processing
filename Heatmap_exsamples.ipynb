Heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
import datetime as dt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/asadbek11211/praktikum_datasets/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]

df.head()
```

```python
df1 = df.groupby(['year','month']).max()

df1 = df1.unstack(level=1)
df1
```

```python
plt.figure(figsize=(18,6))
plt.title("Coming museum")
sns.heatmap(data=df1, cmap='Blues_r', linewidth=1.5, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```
