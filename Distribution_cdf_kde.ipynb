# The distribution is a graphic value in the variable (column) and that provides their likelihood.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

We can see that most of the cars above the above machines are close to $ 20,000.

Let's see the distribution of the kilometer of cars in the table:

```python

sns.displot(data=df, x='mileage', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='mileage', hue='transmission', multiple='stack', height=5, aspect=2 )
plt.show()
```

'**col**' a parameter, each graph can be distinguished:

```python
sns.displot(data=df, x='mileage', col='transmission', multiple='stack', height=5, aspect=2 )
plt.show()
```

### KDE - Kernel Dennity Estismism
Another way to assess the division of values is a graphic that indicates the density of random values.
Full information about KDE: https://en.wikipedia.org/wiki/Kernel_density_Density (also in Russian)

KDE can use the defloside of the schedule:

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

The `Dissolution Distrustion and KDE can be collected:

```python
sns.displot(data=df, x='mileage', kde=True, height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

# kdeplot()

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', shade=True)
plt.xticks(rotation=90)
plt.show()
```

We can add an additional variable using the 'hue':

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='transmission', shade=True)
plt.xticks(rotation=90)
plt.show()
```

`displot` and `hue`

```python
sns.displot(data=df, x='price', hue='fuelType', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', col='fuelType', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

### two-dimensional (2d) KDE

## CDF - Cumulative Distribution Function
The cumulative function function is a parameter of the value given.

Full: https://en.wikipedia.org/wiki/cumulative_distribution_function

```python
sns.displot(data=df, x="price", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both') # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', hue='fuelType', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
